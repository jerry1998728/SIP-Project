```python
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
%matplotlib inline
```

```python
aapl_cul_returns_path = Path('Data/aapl_cumulative_returns.csv')
amzn_cul_returns_path = Path('Data/amzn_cumulative_returns.csv')
fb_cul_returns_path = Path('Data/fb_cumulative_returns.csv')
msft_cul_returns_path = Path('Data/msft_cumulative_returns.csv')
nflx_cul_returns_path = Path('Data/nflx_cumulative_returns.csv')
tsla_cul_returns_path = Path('Data/tsla_cumulative_returns.csv')
wmt_cul_returns_path = Path('Data/wmt_cumulative_returns.csv')
spy_cul_returns_path = Path('Data/spy_cumulative_returns.csv')
```

```python
aapl_cul_returns_csv = pd.read_csv(aapl_cul_returns_path)
aapl_cul_returns_csv = aapl_cul_returns_csv.drop(columns=['Unnamed: 0','mean'])
amzn_cul_returns_csv = pd.read_csv(amzn_cul_returns_path)
amzn_cul_returns_csv = amzn_cul_returns_csv.drop(columns=['Unnamed: 0','mean'])
fb_cul_returns_csv = pd.read_csv(fb_cul_returns_path)
fb_cul_returns_csv = fb_cul_returns_csv.drop(columns=['Unnamed: 0','mean'])
msft_cul_returns_csv = pd.read_csv(msft_cul_returns_path)
msft_cul_returns_csv = msft_cul_returns_csv.drop(columns=['Unnamed: 0','mean'])
nflx_cul_returns_csv = pd.read_csv(nflx_cul_returns_path)
nflx_cul_returns_csv = nflx_cul_returns_csv.drop(columns=['Unnamed: 0','mean'])
tsla_cul_returns_csv = pd.read_csv(tsla_cul_returns_path)
tsla_cul_returns_csv = tsla_cul_returns_csv.drop(columns=['Unnamed: 0','mean'])
wmt_cul_returns_csv = pd.read_csv(wmt_cul_returns_path)
wmt_cul_returns_csv = wmt_cul_returns_csv.drop(columns=['Unnamed: 0','mean'])
spy_cul_returns_csv = pd.read_csv(spy_cul_returns_path)
spy_cul_returns_csv = spy_cul_returns_csv.drop(columns=['Unnamed: 0','mean'])
```

```python
aapl_cul_returns_csv = aapl_cul_returns_csv.rename(columns={'median':'AAPL median',
                                     'min':'AAPL min',
                                     'max':'AAPL max'})
amzn_cul_returns_csv = amzn_cul_returns_csv.rename(columns={'median':'AMZN median',
                                     'min':'AMZN min',
                                     'max':'AMZN max'})
fb_cul_returns_csv = fb_cul_returns_csv.rename(columns={'median':'FB median',
                                     'min':'FB min',
                                     'max':'FB max'})
msft_cul_returns_csv = msft_cul_returns_csv.rename(columns={'median':'MSFT median',
                                     'min':'MSFT min',
                                     'max':'MSFT max'})
nflx_cul_returns_csv = nflx_cul_returns_csv.rename(columns={'median':'NFLX median',
                                     'min':'NFLX min',
                                     'max':'NFLX max'})
tsla_cul_returns_csv = tsla_cul_returns_csv.rename(columns={'median':'TSLA median',
                                     'min':'TSLA min',
                                     'max':'TSLA max'})
wmt_cul_returns_csv = wmt_cul_returns_csv.rename(columns={'median':'WMT median',
                                     'min':'WMT min',
                                     'max':'WMT max'})
spy_cul_returns_csv = spy_cul_returns_csv.rename(columns={'median':'SPY median',
                                     'min':'SPY min',
                                     'max':'SPY max'})
```

```python
all_cul_returns = pd.concat([aapl_cul_returns_csv,
                             amzn_cul_returns_csv,
                             fb_cul_returns_csv,
                             msft_cul_returns_csv,
                             nflx_cul_returns_csv,
                             tsla_cul_returns_csv,
                             wmt_cul_returns_csv,
                             spy_cul_returns_csv],join='inner',axis=1)

all_cul_returns.head()
```
